```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from sklearn.model_selection import train_test_split

from PIL import Image

from pandas_profiling import ProfileReport

sns.set(style='whitegrid', font_scale=1.5)

TARGET = 'label'
RANDOM_STATE = 42
```

```python
def show_values_on_bars(axs, h_v="v", space=0.4):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```

```python
full_train_df = pd.read_csv('train.csv')
full_test_df = pd.read_csv('test.csv')
```

```python
plt.figure(figsize=(10, 5))
g = sns.countplot(data=full_train_df, y='label')
for p in g.patches:
    width = p.get_width()
    g.text(width + 50,
           p.get_y() + p.get_height() / 2,
           '{:1}'.format(width),
            ha = 'left', 
            va = 'center') 
plt.show()
```

```python
f, axes = plt.subplots(10, 10, figsize=(15, 15))
for label, ddf in full_train_df.groupby('label'):
    for idx, row in enumerate(ddf.sample(10).values):
        img = row[1:]
        axes[label, idx].imshow(img.reshape((28, 28)))

[ax.set_axis_off() for ax in axes.ravel()]
plt.show()
```

```python
X_train, y_train = full_train_df.values[:,1:], full_train_df.values[:,0]
X_train = X_train / 255.0

X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=RANDOM_STATE, stratify=y_train)

X_test = full_test_df.values
X_test = X_test / 255.0
```

```python
print('X_train.shape =', X_train.shape)
print('X_valid.shape =', X_valid.shape)
print('X_test.shape =', X_test.shape)
```

```python
model = tf.keras.Sequential([
        tf.keras.layers.Reshape(input_shape=(784,), target_shape=(28, 28, 1)),

        tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
        tf.keras.layers.BatchNormalization(axis=-1),

        tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(),
        tf.keras.layers.BatchNormalization(axis=-1),

        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.BatchNormalization(axis=-1),

        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(),

        tf.keras.layers.Flatten(),
        tf.keras.layers.BatchNormalization(),

        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        
        tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.summary()
```

```python
%%time

model.fit(X_train, y_train, batch_size=128, epochs=2, validation_data=(X_valid, y_valid))
```

```python
valid_loss, valid_acc = model.evaluate(X_valid, y_valid, verbose=2)

print('Valid accuracy:', valid_acc)
```

```python
predictions = model.predict(X_test)
y_pred = np.argmax(predictions, axis=-1)

submission_df = pd.DataFrame(data={
    'ImageId': np.arange(1, len(full_test_df) + 1),
    TARGET: y_pred
})
submission_df.to_csv('output.csv', index=False)
submission_df
```
